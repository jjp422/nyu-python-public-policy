```
def graffiti(row): #indents will throw this off?? How to prevent this?
  if pd.notnull(row.descriptor):
    if 'graffiti' in row.descriptor:
      return 'True' 
    elif 'graffiti' in row.complaint_type:
      return 'True' 
    else:
        return 'False'     
```

```
type(graffiti)
```

```
df['graffiti_flag'] = df.apply(graffiti, axis=1)
```
